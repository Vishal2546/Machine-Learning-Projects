```python
#Importing required packages
#Our primary packages for this project are going to be pandas for data processing, 
#NumPy to work with arrays, 
#matplotlib & seaborn for data visualizations, 
#and finally sk-learn for building an evaluating our ML model. 
#Let’s import all the required packages into our python environment.
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as pt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Importing Data
#As I said before, we are going to work with the house price dataset that contains various features
#and information about the house and its sale price. 
#Using the ‘read_csv’ function provided by the Pandas package, we can import the data into our python environment. 
#After importing the data, we can use the ‘head’ function to get a glimpse of our dataset.
df=pd.read_excel('DS - Assignment Part 1 data set.xlsx')
df
```

```python
#checking null sets
df.isnull().sum()
```

```python
#schema check
df.info()
```

```python
#taking transaction date as a type of string
df['Transaction date']=df['Transaction date'].astype(str)
```

```python
#split the .0 part
df['Transaction date']=df['Transaction date'].apply(lambda x: x.split('.')[0])
```

```python
df
```

```python
#again taking txn date as a integer type
df['Transaction date']=df['Transaction date'].astype(int)
df
```

```python

df.info()
```

```python
#house age as a type of integer
df['House Age']=np.ceil(df['House Age']).astype(int)
```

```python
#Distance from nearest Metro station (km) is also as in form of integer type
df['Distance from nearest Metro station (km)']=df['Distance from nearest Metro station (km)'].astype(int)
df
```

```python
#for calculating total price of house= size of house in square feet multiply with price of per unit area
df['Total Price of House']= df['House size (sqft)'] * df['House price of unit area']
df
```

```python
df.describe()
```

```python
# The sns is short name use for seaborn python library.
#The heatmap especially uses to show 2D (two dimensional) data in graphical format.
#Heatmaps are very useful to find relations between two variables in a dataset.
#Heatmap can be easily produced using the ‘heatmap’ function provided by the seaborn package in python.
sns.heatmap(df.corr(),annot=True)
```

```python
#separate the independent values into X and depended value into y
x=df.iloc[:,:-1]
y=df['Total Price of House']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
# create the LinearRegression() model for training
from sklearn.linear_model import LinearRegression
Lr=LinearRegression()
Lr.fit(xtrain,ytrain)
ypred=Lr.predict(xtest)
```

```python
from sklearn.metrics import r2_score
```

```python
#check r2_score
#Model Evaluation
r2_score(ytest,ypred)
```

```python
Lr.score(xtrain,ytrain)
```
